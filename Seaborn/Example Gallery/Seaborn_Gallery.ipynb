## Gallery

### Anscombe's quarter

```python
import seaborn as sns
sns.set_theme(style="ticks")
df = sns.load_dataset("anscombe") # Dataset 1
df.info()
```

```python
df.head()
```

```python
# Show the results of a linear regression within each dataset
sns.lmplot(
    data=df,
    x="x",
    y="y",
    col="dataset",
    hue="dataset",
    col_wrap=2,
    palette="muted",
    ci=None,
    height=4,
    scatter_kws={"s":50, "alpha":1}
)
```

### Scatterplot with multiple semantics



```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")
diamonds = sns.load_dataset("diamonds")
```

```python
diamonds.info()
```

```python
diamonds.head()
```

```python
# Draw a scatter plot while assigning point colors and sizes to different variables in a dataset
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f,
             left=True,
             bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(
    data=diamonds,
    x="carat",
    y="price",
    hue="clarity",
    size="depth",
    palette="ch:r=-.2,d=.3_r",
    hue_order=clarity_ranking,
    sizes=(1, 8),
    linewidth=0,
    ax=ax
)
```

### Line plots on multiple facets

### Grouped barplots


```python
import seaborn as sns
sns.set_theme(style="whitegrid")
penguins = sns.load_dataset("penguins") # Used Earlier
```

```python
# Draw a nested barplot by species and sex
g = sns.catplot(
    data=penguins,
    kind="bar",
    x="species",
    y="body_mass_g",
    hue="sex",
    errorbar="sd",
    palette="dark",
    alpha=.6,
    height=6
)
g.despine(left=True)
g.set_axis_labels("", "Body mass (g)")
g.legend.set_title("") # This is a useful way to remove title from legend
```

### Grouped Boxplots


```python
import seaborn as sns
sns.set_theme(style="ticks", palette="pastel")
tips = sns.load_dataset("tips")
```

```python
tips.info()
```

```python
# Draw a nested boxplot to show bills by day and time
sns.boxplot(
    data = tips,
    x="day",
    y="total_bill",
    hue="smoker",
    palette=["m", "g"],
)
sns.despine(offset=10, trim=True)
```

### Grouped violinplots with split violins



```python
import seaborn as sns
sns.set_theme(style="dark")
tips = sns.load_dataset("tips") # Used earlier
```

```python
# Draw a nested violin plot and split the violins for easier comparison
sns.violinplot(
    data = tips,
    x="day",
    y="total_bill",
    hue="smoker",
    split=True,
    inner="quart",
    fill=False,
    palette={"Yes": "g", "No": ".35"}
)
```

### Scatterplot heatmap


```python
import seaborn as sns
sns.set_theme(style="whitegrid")

# Load the brain networks dataset, select subset, and collapse the multi-index
df = sns.load_dataset("brain_networks", header=[0, 1, 2], index_col=0)

used_networks = [1, 5, 6, 7, 8, 12, 13, 17]
used_columns = (df.columns
                  .get_level_values("network")
                  .astype(int)
                  .isin(used_networks))
df = df.loc[:, used_columns]

df.columns = df.columns.map("-".join)

# Compute a correlation matrix and convert to long-form
corr_mat = df.corr().stack().reset_index(name="correlation")

# Draw each cell as a scatter point with varying size and color
g = sns.relplot(
    data=corr_mat,
    x="level_0", y="level_1", hue="correlation", size="correlation",
    palette="vlag", hue_norm=(-1, 1), edgecolor=".7",
    height=10, sizes=(50, 250), size_norm=(-.2, .8),
)

# Tweak the figure to finalize
g.set(xlabel="", ylabel="", aspect="equal")
g.despine(left=True, bottom=True)
g.ax.margins(.02)
for label in g.ax.get_xticklabels():
    label.set_rotation(90)
```
